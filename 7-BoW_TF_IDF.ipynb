```python
data = ["AI acts as the umbrella that enables machines to perform tasks requiring human-like understanding and reasoning.",
"Machine Learning allows systems to automatically learn patterns from data and improve their performance over time.",
"Deep Learning extends ML by using multi-layer neural networks to capture complex structures in large-scale datasets.",
"In Natural Language Processing, these methods help computers analyze, understand, and generate human language.",
"Text vectorization transforms raw text into numerical formats such as TF-IDF scores or dense embeddings so models can process it."]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

count_vec = CountVectorizer()
count_vec_fit = count_vec.fit_transform(data)

bag_of_words = pd.DataFrame(count_vec_fit.toarray(), columns = count_vec.get_feature_names_out())
print(bag_of_words.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec = TfidfVectorizer()
tfidf_vec_fit = tfidf_vec.fit_transform(data)

tfidf_bag = pd.DataFrame(tfidf_vec_fit.toarray(), columns=tfidf_vec.get_feature_names_out())
print(tfidf_bag)
```
