```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

```python
data = pd.read_csv("10. tripadvisor_hotel_reviews.csv")
data.head()
```

```python
data['Review_process'] = data['Review'].str.lower()
data.head()
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
data['Review_process'] = data['Review'].apply(lambda x : ' '.join([word for word in x.split() if word not in en_stopwords]))
data.Review_process[0]
```

```python
data['Review_process'] = data.apply(lambda x : re.sub(r'[*]', ' star', x['Review_process']), axis=1)
data['Review_process'] = data.apply(lambda x : re.sub(r'[^\s\w]', '', x['Review_process']), axis=1)
```

```python
data.head()
```

```python
data['Review_tokens'] = data.apply(lambda x : word_tokenize(x['Review_process']), axis=1)
```

```python
data.head()
```

```python
ps = PorterStemmer()
data['stemm_review'] = data['Review_tokens'].apply(lambda tokens : [ps.stem(token) for token in tokens])
```

```python
lm = WordNetLemmatizer()
data['lemmatizer_review'] = data['Review_tokens'].apply(lambda tokens : [lm.lemmatize(token) for token in tokens])
```

```python
data.head()
```

```python
clean_tokens = sum(data['lemmatizer_review'],[])
unigram = pd.Series(nltk.ngrams(clean_tokens ,1)).value_counts()
bigram = pd.Series(nltk.ngrams(clean_tokens ,2)).value_counts()
print(unigram[:5])
print(bigram[:5])
```
